# Solution of Lotka-Volterra Equation
$$ \dot{x} = \alpha x - \beta xy $$
$$ \dot{y} = \delta xy - \gamma y$$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

$${S} = \begin{bmatrix}
\dot{x}\\
\dot{y}
\end{bmatrix} = \begin{bmatrix}\alpha x - \beta xy \\
\delta xy - \gamma y\end{bmatrix}$$

```python
alpha=1.1
beta=0.4
gamma=0.4
delta=0.1
t = np.linspace(0,100,10000)

def dSdt(t,S):
    x , y = S
    return np.array([alpha*x-beta*x*y,
              delta*x*y-gamma*y])
plt.grid(True)
plt.title("Lotka Volterra Equation")
for i in range(5):
    y0 = [0.1*i,0.2*i]
    sol = odeint(dSdt,y0=y0,t=t,tfirst=True)
    plt.subplot(212)
    plt.plot(sol[:,0],sol[:,1])

plt.subplot(211)
plt.plot(t,sol[:,0])
plt.plot(t,sol[:,1])
plt.show()
```
